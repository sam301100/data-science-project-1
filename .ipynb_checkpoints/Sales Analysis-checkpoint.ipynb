## Data Cleaning

```python
#To Drop rows with 'NaN' values

#Find rows with Nan values
nan_df = all_data[all_data.isna().any(axis=1)]

nan_df
```

```python
#Drop the rows where all elements are missing.

all_data = all_data.dropna(how='all')
all_data
```

### Add a Month column

```python
#Augment data with additional columns 
#i.e. to find the best month for sales, add a column of month


all_data['Month'] = all_data['Order Date'].str[0:2]          #since the date is a string
```

```python
#Locate months with 'Or' values
all_data.loc[all_data['Month'] == 'Or']
```

```python
#remove rows with value 'Or' in the month column

all_data = all_data[all_data.Month != 'Or']
```

## Task 1 : What was the best month for sales? How much was earned that month?

```python
#Convert the month values to int
all_data['Month'] = all_data['Month'].astype('int32')
```

```python
#Calculate Sales

all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('float32')
all_data['Price Each'] = all_data['Price Each'].astype('float32')

#Adding sales column
all_data["Sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]

all_data
```

```python
#Individual Maximum sale
all_data[all_data.Sales == all_data.Sales.max()]
```

```python
results = all_data.groupby('Month').sum()
results
```

```python
#Finding the maximum sales month among all the months

max_sales_month = results[results.Sales == results.Sales.max()]
max_sales_month
```

```python
max_sales_month['Sales']
```

```python
months = range(1,13)

figure(figsize=(8, 6), dpi=70)

plt.bar(months, results['Sales'])

plt.xticks(months)
plt.yticks(results['Sales'])
plt.xlabel('Month Number', size=15)
plt.ylabel('Sales in USD $', size=15)

plt.show()
```

## Task 2 : Which city had the highest number of sales?

### Add a City column

```python
#Using .apply()

#all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

#all_data.head()


#OR


def get_city(address):
    return address.split(',')[1]


def get_state(address):
    return address.split(',')[2].split(" ")[1]
    
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

all_data.head()
```

```python
result = all_data.groupby('City').sum()
result
```

```python
max_sales_city = result[result.Sales == result.Sales.max()]
max_sales_city
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, result['Sales'], color='maroon')

plt.xticks(cities, rotation='vertical')
plt.yticks(result['Sales'])
plt.xlabel('City Names')
plt.ylabel('Sales in USD $')

plt.show()
```

## Task 3 : What time should we display advertisemens to maximize the likelihood of customer’s buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute

all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.grid()
plt.show()


#From the below graph, advertisements should be posted around - 11 am or 7 pm
```

## Task 4 : What products are most often sold together?

```python
all_data['Order ID'].duplicated(keep = False)
```

```python
#Creating new dataframe with only the rows having their duplicates present (duplicate order ID)

df = all_data[all_data['Order ID'].duplicated(keep = False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))

df.head(10)
```

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
    
print(count)
```

```python
count.most_common(10)
```

## Task 5 : Which product was sold the most? Why do you think it sold the most?

```python
all_data.head()
```

```python
group_of_products = all_data.groupby('Product')
quantity_ordered = group_of_products.sum()['Quantity Ordered']



print(group_of_products.sum()['Quantity Ordered'])
```

```python
products = [product for product, df in group_of_products]

plt.bar(products, quantity_ordered)
plt.xlabel('Product')
plt.ylabel('Number Ordered')
plt.xticks(products, rotation='vertical')
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

print(prices)
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered)
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical')
plt.show()
```
